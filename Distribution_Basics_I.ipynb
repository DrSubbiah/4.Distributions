# <font color = maroon> Introduction

Mathematical description of any event involves a basic notion called **Functions**. They are fundamentally provide a symbolic way of representing possible (theoretical / hypothetical / intuitional) relationships. Mostly,a variable or quantity of interest (QoI) will be related to one or more variables that could influence its behaviour. The former (QoI) is termed as dependent variable and the latter (influencers) are independent variables.  In fact, from the form of a function it could be possible to understand different characteristics of the dependent variable (QoI) such as shape, spread, maximum etc. Other mathematical principles from arithmetic operations $(+, -,\times,\div)$ to calculus ideas also help this process. This notes provides a basic idea about different form of relations that could happen between variables or quantities. Some important and widely used terms and terminologies are also discussed with suitable examples and illustrations with graphical outputs from R.

# <font color = maroon> Functional Forms 

Symbolically, a function can be written as $$y=f(x)$$ where $x$ is (one or more) independent and $y$ is the dependent variable; $f$ stands for a relation between these two variables. Needless to say, these are arbitrary symbols and can be replaced by any contextual symbols such as $u=g(t), \theta=f_1(r)$ and etc. There can be infinitely many ways to define mathematical forms; however, it may be understood that a relation between variables is to explain the way one *influences* the other. In other words, this means the way a variable (dependent) *changes or not* when the other (independent) changes. ## <span style="color:maroon"> Five Forms 

The above five types are mainly based on the direction (growth / decay) and the speed of change that could happen in defining a relation between variables. Following functions will capture these features in two major measurement methods, Linear and Angular. 

In the subsequent contents, $x$ is an independent variable (scalar or vector); $k$ is a constant specified according to the necessity in defining a function

1. **Constant Function** $f(x)=k$

1. **Power Function** $f(x) = ax^b+c$, except possibly at $x=0$

    * base is a variable
    
    * exponent (power) is a constant

1. **Exponential Function** $f(x) = k^{x}$ with $k>0$; in particular $k=e$ draws much scientific attention

    * base is a constant 
  
    * exponent (power) is a variable

1. **Logarithmic functions** $f(x) = \log_{m}x$ where $a>0$ is the base  
    * Common logarithm: $m=10$
   
    * Natural logarithm: $m=e$, the Euler's number that lies between 2 and 3 (2.71, approximated to two decimals)

## A note on Euler's Number - e

This number can be understood as a limiting value of a function $f(x)=(1+\frac{1}{x})^x,~x>0$; that is, this function approaches the value of $e$ when $x$ increases . Following figure illustrates this 

Euler e.png

The constants $a,b,c$ are called **parameters** which are unknown constants and the corresponding expression is representing a family of functions; for example $ax+c, (b=1)$ represents a family of straight line and with $c=0$ it is a family of straight line passing through origin. However, if all the parameter values are known then the resultant function is a member of that family of function. For example $f(x)=x$ and $f(x)=-x$ both are members of straight line family passing through origin but differ in the direction

## Linear and Non-Linear Functions

Power functions $f(x)=x^k$ with $k > 2$ are called Non-Linear or Curvilinear functions

When $k=1$ it is called a Linear Function or straight line. The major difference is the rate of change of $y$ with respect to $x$ is constant in linear functions compared to non-linear functions.
