# <font color = maroon> Probability Function

In general a distribution is attempting to represent a random quantity $X$ whose outcomes are unpredictable in nature but could be understood by its ***Range*** $\mathscr{A}_x$ (possible values, it can assume). This  mathematical form depicts a possible behavior of the random variable in terms of its shape and spread over the range $\mathscr{A}_x$. Further such function is useful to compute the uncertainty in terms of probability; hence, it has to satisfy the natural two conditions involved with the notion of probability 

1. Non negativity

1. Total Probability = 1

Since $\mathscr{A}_x$  is the exhaustive possible values of $X$ the above two conditions can be related typically to the values on $\mathscr{A}_x$. If $f(x)$ is such function of a random variable $X$, then the above two conditions are 

1. $f(x) \ge 0$ (equality holds for all $x$ **not** in $A_x$)

1. $\int_{\mathscr{A}_x}f(x)~~dx=1$ or $\sum_{\mathscr{A}_x}f(x)=1$

Note that summation or integration over $\mathscr{A}_x$ depends on the nature of the variable $X$ as Discrete or Continuous respectively. Discrete set $\mathscr{A}_x$ contains finite or countably infinite values $\{x_1,x_2,x_3,\cdots\cdots x_n\}$ or $\{x_1,x_2,x_3,\cdots\cdots \cdots\}$; on the other hand,  $\mathscr{A}_x$ is an uncountable set over the real numbers and depends on the dimension of $X$. If $X$ is a one dimensional continuous variable then its range $\mathscr{A}_x$ could be a bounded  interval $(a,b)$ (a and b are finite real numbers with $a < b$) or a non-negative interval $[0, \infty)$ or complete real line  $(-\infty, \infty)$. More importantly, the various form of the function $f(x)$ is characterized by parameters involved in the functional form. Also, $f(x)$ is called **Probability Mass Function (PMF)** or  **Probability Density Function (PMF)** respectively when $X$ is a discrete or a continuous variable


- **<font color = darkblue> Summarizing, in the language of Mathematical Statistics, $$X \sim f(X|\theta)$$ implies that X is random variable with PDF $f$ with parameter $\theta$** 
   
   - $\theta$ can be any dimension
   
   - Range of X is denoted as $\mathscr{A}_x$
   
   - $|$ **should** be read as given
   
   - $f(X|\theta)$ should be read as, "probability mass (density) function of x **given** $\theta$ is equal to ........"
   
   - Whenever $\theta$ is known, $f(x|\theta)$ computes value of the function at $x$ of $X$

# <font color = maroon> Reading a Theoretical distribution

A theoretical distribution has practical reasoning so as to define in a functional form as described earlier. Following three are just to exemplify the basic requirements one has to know about a distribution. Genesis of many standard theoretical distributions (and other features) may be referred [here](https://www.amazon.ae/Continuous-Univariate-Distributions-Norman-Johnson/dp/0471584940); other similar texts by these authors are also worth reading.

**<font color = darkgreen>Normal distribution**

A Normal random variable is a continuous random variable of one dimension that illustrates the a family of distribution with two parameter    

- PDF: $\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

- Parameters: Location: $\mu$, Scale: $\sigma$ 

- Range of X, $\mathscr{A}_x$: $-\infty \lt x \lt\infty$

- Parameter Space: $-\infty\lt\mu\lt\infty$, $\sigma\gt0$

**<font color = darkgreen>Beta distribution**

This is an example for a univariate distribution which has only a shape parameter

- PDF: $\frac{1}{\textrm{Beta}(\alpha,\beta)}x^{(\alpha-1)}(1-x)^{(\beta-1)}$
  
  * $\textrm{Beta}$ is the beta function $\textrm{Beta}(a,b)=\frac{\Gamma{a}\Gamma{b}}{\Gamma{a+b}}$ and $\Gamma{a}$ is the gamma function defined as $\Gamma{a}=\int_0^\infty e^{-t}~~t^{a-1}~~dt$
  
- Parameters: Shape: $\alpha,\beta$

- Range of X, $\mathscr{A}_x$: $0\lt x\lt1$

- Parameter Space: $\alpha\gt0$,$\beta\gt0$

**<font color = darkgreen>Bernoulli distribution**

This is a classic example for a discrete variable with binary outcomes, that are mutually exclusive in nature. For example, the two possible values could be  "Success or Failure" or "Yes or No" or $\{0,1\}$ with $0$ and $1$ denote the two states of the discrete binary variable

- PMF: $\theta^x (1-\theta)^{1-x}$

- Parameters: Shape: $\theta$

- Range of X, $\mathscr{A}_x$: $x = 0,1$

- Parameter Space: $0\lt \theta \lt 1$

# <font color = maroon> Probability computations

## <font color = blue> Event Probabilites

Let a and b be two real numbers. Let $X$ be a **Continuous random variable** with range $\mathscr A_x$ and **PDF** $f(x)$. Recall,

1. $f(x) \ge 0$ for all $x$ in $A_x$

1. $\int_{\mathscr A_x}f(x)~~dx=1$ or $\sum_{\mathscr A_x}f(x)=1$

Let $L, U$ be chosen properly from the range $\mathscr A_x$ of $X$

- $X$ is at most $b$: $p(X<b)=p(X\le b)=\int_L^bf(x)~dx$ 

- $X$ is at least $a$: $p(X>a) = p(X\ge a)=\int_a^Uf(x)~dx$ 

- $X$ lies between $a$ and $b$: $p(a\le X\le b)=p(a\le X\lt b)=p(a\lt X\le b)=p(a\lt X\lt b)=\int_a^bf(x)~dx$ 

- $P(X=b) = 0$ 

***<font color = red> These formulas will be helpful to distinguish between value of the function $f(X|\theta)$ computed at a point and probability computations associated with a continuous random variable $X$***. For example, consider a Uniform distribution in $\mathscr {A}_x=(a_1, a_2)$ where $a_1, a_2$ are two real numbers such that $a_1< a_2$. Its PDF is $f(x)=\frac{1}{a_2-a_1}$ whenever $a_1 < x < a_2$. It can be understood that $f(x)$ is constant, independent of x. However, to find probability associated with $X$ one has to integrate the function $f(X)$ in the given range of $X$ 

As an illustration, consider $f(x)=\frac{1}{0.5-0}=2$ in the range $0<x<0.5$. Hence, **value** of $f(x)$ is 2 for all values in $(0, 0.5)$; however, probabilities associated with $X$ is calculated using integration of $f(x)$ and restricted to be within [0,1]


In the case of a **discrete random variable**,

- integration is replaced with summation in the probability computations

- mass function $P(X=x)$ yields probability that $X$ takes on a specific value $x$ in contrast to a continuous random variable

- $P(X=b)$ may be non-zero

- $<$ or $>$ are different from $\le$ or $\ge$ in the computations because the former does not include the end points. For example if a discrete random variable has a range $\mathscr{A}_x=\{0,1,2,\cdots\}$, then $P(X\ge0)$ includes $0$ whereas $P(X\gt0)$ will not include $0$

## <font color = blue> Cummulative Density Function (CDF) 

CDF or ***Distribution Function*** computes probabilities $$F_X(b)=p(X\le b) = \int_{-\infty}^b f(x)~dx$$ or 

$$F_X(b)=P[X \le b] = \sum_{-\infty}^bf(x)$$

It should be noted that 

1. $b$ can be any real number

    - this formula is applicable even when $b$ lies outside $\mathscr{A}_x$, the range of X

1. $0\le F_X(b) \le 1$

## <font color = blue> Quantile Function QF or Inverse-CDF

Quantiles (aka percentiles) are defined as points of divisions for a distribution. Scaled from 0 (minimum) to 100 (maximum) or 0 to 1. It indicates the percentage of values (of the variable) that fall below a particular value. 

For some wide spread applications, see [here](https://www.princetonreview.com/grad-school-advice/good-gre-scores) and [here](https://engineering.careers360.com/articles/jee-main-marks-vs-percentile)

This is better than percentage because it gives the relative position of a value (of the variable). Percentile is a  value $x_\alpha$ of the variable $X$ such that probability that $X$ will be less than or equal to $x_\alpha$ is $\alpha$. symbolically,$100\times\alpha^{th}$ percentile is 

$$p[X \le x_{\alpha}]=\alpha$$

where $0 \le \alpha \le 1$

For example, we assume income distribution (a Gamma distribution with shape 100 and scale = 0.1) and want to know the value for picking top income group, say 10%. Then we need to find $90^{th}$ percentile. Less than that value (income), 90% of the income lie less than $x_{\alpha}$; equivalently 10% are more than $x_{\alpha}$. 


Quantile.png

For income example,$\alpha = 0.90$ and $x_{\alpha}$ is 2.64  (Please refer graph)


These computational expressions may bring a fact that **quantiles**  can be understood as inverse function of **CDF** from the expression $\alpha = p[X\le b]$.

* Distribution function computes $\alpha$ when $b$ is given and $b$ can be any real number

* Quantiles gives $b$ when probability $\alpha$ is known

## <font color = blue> Summaries of a distribution

Assume the variable is a quantitative (continuous) variable, then its range is an interval in real line $\mathscr R=(-\infty, +\infty)$.

Let $f(x)$ be a pdf of $X$ in the range $\mathscr A_x \subset \mathscr R$

Further assume any function of X, say $g(X)$. We define expected value of $g(X)$ or expectation of $g(X)$ as

$$E[g(X)] = \int_{\mathscr {A}_x} {g(x)f(x)} dx$$  

Specifically, if $g(X) = X$ we get mean

$$E[X] =\int_{\mathscr{A}_x}{xf(x)} dx$$

Let $g(X)=X^2$. $$E[X^2] = \int_{\mathscr{A}_x}{x^2f(x)} dx$$

Then  we use this to get the variance of X as 

$$V(X)=E[X^2]-E[X]^2$$
