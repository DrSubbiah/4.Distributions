# <font color="darkblue"> Introduction
Random variable and theoretical distributions play very important role in modeling uncertainty. It offers a syntactic approach for the language of uncertainty

The mathematical form for a distribution is a function $f(x)$ satisfying two conditions;

  - Non-negativity  $f(x) \ge 0$
  
  - Total probability is 1 $\sum f(x) = 1$ or $\int f(x)~dx = 1$

## <font color="maroon"> Few important keywords

- Random variable

- **Range** of a random variable

- **Nature** of a variable based on its range
  
  - Discrete / Continuous
  
  - Bounded / Unbounded (Finite / Infinite)

- Probability **Mass** / **Density** Functions (PMF / PDF)

- **Summarizing** a distribution

  - Expectation of functions of a random variable
  
  - Probability statements about a random variable
  
  - Order statistics of a random variable

- Nature and Role of parameters

# <font color="maroon"> Conventional Notations

Following is a broad way of understanding various symbols used to represent random variables and distributions. However, this can be modified in a specific context; for more details, please [refer](https://en.wikipedia.org/wiki/List_of_mathematical_symbols_by_subject)

- Variables: Upper case Latin alphabets $X, Y, Z$ (preferably last 6 letters)

- Unknown Value of a variable: Lower case Latin alphabets $x, y, z$ (preferably last 6 letters)

- Constants: Lower case Latin alphabets $a, b, c$ (preferably first 5 letters)

- Parameters: Lower case Latin alphabets $a,b,c,k,l, m, n$ (except possibly the last 5 letters) or Lower case Greek alphabets $\alpha, \beta, \gamma$

- Also, Upper case Latin alphabets are used to list more variables $X_1, X_2, \cdots\cdots X_n$ and values of corresponding variables are denoted using lower case Latin alphabets $x_1, x_2, \cdots\cdots x_n$ are used.

- In the previous point, we could note another aspect called ***<span style="color:darkgreen">index*** to specify the number of variables.Lower Latin alphabets are used for index such as $i=1,2,3\cdots\cdots N$ or $j=1,2,3\cdots\cdots n$

- **<font color="darkblue">Collectively, in the language of Mathematical Statistics,  $$X \sim f(X|\theta)$$ implies that X is random variable with PDF $f$ with parameter $\theta$**
   
   - $\theta$ can be any dimension
   
   - Range of X is denoted as $\mathscr{A}_X$
   
   - $|$ **should** be read as given
   
   - $f(X|\theta)$ should be read as, "probability mass (density) function of x **given** $\theta$ is equal to ........"
   
   - Whenever $\theta$ is known, $f(x|\theta)$ computes value of the function at $x$ of $X$

# <font color="darkblue"> Role of Parameters - Learning from Basic Function

Each of the five functions exhibit interesting characteristics for suitable values of respective parameters. In statistical paradigm three types of parameters are more prominent; ***<span style="color:darkblue">location, shape, and scale*** parameters.

```python
import numpy as np
import pandas as pd
import scipy as sc
from scipy import stats as st
import math
import matplotlib.pyplot as plt
```

# <font color="darkblue"> Location Parameter

```python
x= np.linspace(-5, 5, 100)

def f1(x):
    return x ** 2
f12 = np.vectorize(f1)

def f2(x):
    return (x-3)**2
f22 = np.vectorize(f2)

def f3(x):
    return (x+3)**2
f32 = np.vectorize(f3)

fig = plt.figure(figsize = (20, 8))

plt.plot(x, f12(x),label="$x^2, a=0$",color="magenta")
plt.plot(x, f22(x),label="${(x-3)}^2,a=3$")
plt.plot(x, f32(x),label="${(x+3)}^2, a=-3$")

plt.title('Location Parameter $(x-a)^k$',fontsize=20)
plt.legend(fontsize=15)

plt.axhline(y=0, color='g', linestyle='--')
plt.axvline(x=0, color='g', linestyle='--')

plt.show()
```

# <font color="darkblue"> Scale Parameter

For different values of scale parameter $b$ in $e^{-bx}$ from 0 to 2

```python
x= np.linspace(0, 2, 100)

def f1(x):
    return math.exp(-x)
f12 = np.vectorize(f1)

def f2(x):
    return math.exp(-2*x)
f22 = np.vectorize(f2)

def f3(x):
    return math.exp(-0.1*x)
f32 = np.vectorize(f3)

fig = plt.figure(figsize = (20, 8))

plt.plot(x, f12(x),label="$e^{-x}, b=1$",color="magenta")
plt.plot(x, f22(x),label="$e^{-2x}, b=2$")
plt.plot(x, f32(x),label="$e^{-0.1x}, b=1/10$")

plt.title('Location Parameter $e^{-bx}$',fontsize=20)
plt.legend(fontsize=15)


plt.show()
```

# <font color="darkblue"> Scale and Location Parameters

Some examples for combination of scale and location parameters, $\frac{1}{b}f(\frac{(x-a)}{b})$. In the following, first curve has scale = 1 to compare when $b > 1 ~\mathrm{or}~ < 1$


```python
x= np.linspace(-9, 7, 100)

def f1(x):
    return math.exp(-x**2)
f12 = np.vectorize(f1)

def f2(x):
    return (1/3)*math.exp(-((x-2)**2)/3)
f22 = np.vectorize(f2)

def f3(x):
    return (1/0.1)*math.exp(-(1/0.1)*(x+2)**2)
f32 = np.vectorize(f3)

fig = plt.figure(figsize = (20, 8))

#plt.plot(x, f12(x),label="$a=0, b=1$",color="magenta")
plt.plot(x, f22(x),label="$a=2, b=3$")
# plt.plot(x, f32(x),label="$a=-2, b=1/0.1$")

plt.title('Scale Location Parameter $e^{-(x-a)^2/b}$',fontsize=20)
plt.legend(fontsize=15)


plt.show()
```

# <font color="darkblue"> Shape Parameter

Few more examples for combination of shape and location parameters

```python
x= np.linspace(0.00001,0.99999, 100)

def f1(x):
    return (1-x)**2
f12 = np.vectorize(f1)

def f2(x):
    return (1-x)**3
f22 = np.vectorize(f2)

def f3(x):
    return (x**-0.5)*(1-x)**-0.5
f32 = np.vectorize(f3)

fig = plt.figure(figsize = (20, 8))

# plt.plot(x, f12(x),label="$a=0, b=2$",color="magenta")
# plt.plot(x, f22(x),label="$a=0, b=3$")
plt.plot(x, f32(x),label="$a=-0.5, b=-0.5$")

plt.title('Scale Location Parameter $x^a{(1-x)}^b$',fontsize=20)
plt.legend(fontsize=15)
```

# <font color="darkblue"> Reading Theoretical Distributions

A theoretical distribution has practical reasoning so as to define in a functional form as described earlier. Following three are just to exemplify the basic requirements one has to know about a distribution.

Genesis of many standard theoretical distributions (and other features) may be referred [here](https://www.amazon.ae/Continuous-Univariate-Distributions-Norman-Johnson/dp/0471584940); other similar texts by these authors are also worth reading.

**<span style="color:darkgreen">Normal distribution**

- PDF: $\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

- Parameters: Location: $\mu$, Scale: $\sigma^2$

- Range of X, $\mathscr{A}$: $-\infty \lt x \lt\infty$

- Parameter Space: $-\infty\lt\mu\lt\infty$, $\sigma^2\gt0$

**<span style="color:darkgreen">Beta distribution**

- PDF: $\frac{1}{\beta(\alpha,\beta)}x^{(\alpha-1)}(1-x)^{(\beta-1)}$

- Parameters: Shape: $\alpha,\beta$

- Range of X, $\mathscr{A}$: $0\lt x\lt1$

- Parameter Space: $\alpha\gt0$,$\beta\gt0$

**<span style="color:darkgreen">Binomial distribution**

- PDF: ${n \choose x} \theta^x (1-\theta)^{n-x}$

- Parameters: Shape: $n,\theta$

- Range of X, $\mathscr{A}$: $x = 0,1,2,...,n$

- Parameter Space: $0\lt \theta \lt1$, $n\gt0$ integer
